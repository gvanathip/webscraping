```python
import pandas as pd
import matplotlib.pyplot as plt
pd.set_option('display.max_colwidth', -1)
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('Keyword-search.xlsx')
```

```python
df.head(3)
```

```python
selected = df.set_index('ข้อความค้นหายอดนิยม')['การแสดงผล'].to_dict()
```

```python
wordcloud = WordCloud(font_path='/content/THSarabunNew.ttf',background_color ='white',max_words=100).fit_words(selected)
fig, ax = plt.subplots(1, 1, figsize=(16, 12))
ax.imshow(wordcloud, interpolation='bilinear')
ax.axis("off")
fig.show()
```
