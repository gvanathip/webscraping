```python
from selenium import webdriver
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome('chromedriver',chrome_options=chrome_options)
driver.get(url)
```

```python
element = driver.find_element_by_xpath('//*[@id="kgrData"]/table/tbody/tr[8]/td[4]/div/div/a')
```

```python
file_path = 'https://population.un.org'+element.get_attribute('pathname')
```

```python
df_1 = pd.read_excel(file_path, index_col=0, sheet_name='ESTIMATES') 
```

```python
df_1 = df_1[15:]
```

```python
new_header = df_1.iloc[0] #grab the first row for the header
df_1 = df_1[1:] #take the data less the header row
df_1.columns = new_header #set the header row as the df header
```

```python
df_1
```

```python
df_2 = pd.read_excel(file_path, index_col=0, sheet_name='MEDIUM VARIANT') 
```

```python
df_2 = df_2[15:]
```

```python
new_header = df_2.iloc[0] #grab the first row for the header
df_2 = df_2[1:] #take the data less the header row
df_2.columns = new_header #set the header row as the df header
```

```python
df_2.head()
```

```python
df_1.columns
```

```python
df_2.columns
```

```python
df_1 = df_1.set_index(list(df_1.columns)[:6])
df_2 = df_2.set_index(list(df_2.columns)[:6])
```

```python
df_1.columns
```

```python
df_2_col = list(set(df_2.columns) - set(df_1.columns))
```

```python
df_2_col.sort()
```

```python
df_2 = df_2[df_2_col]
```

```python
df_1 = df_1.reset_index()
df_2 = df_2.reset_index()
```

```python
df_1 = df_1.drop(columns=['Variant'])
df_2 = df_2.drop(columns=['Variant'])
```

```python
list(df_1.columns)[:5]
```

```python
df = pd.merge(df_1, df_2, on=list(df_1.columns)[:5])
```

```python
df = df.set_index(list(df.columns)[:5])
```

```python
df = pd.DataFrame(df.stack(0))
```

```python
df = df.rename(columns={0:"Total population (thousands)"})
```

```python
df = df.reset_index()
```

```python
df = df.rename(columns={0:"Total population (thousands)","Index":"Year"})
```
